```python
import ast
import numpy as np

# Fungsi untuk mengonversi string menjadi list
def str_to_list(val):
    if val is None or (isinstance(val, float) and np.isnan(val)):
        return []
    
    if isinstance(val, list):
        return val
    
    if isinstance(val, (np.ndarray, pd.Series)):
        return val.tolist()
    
    if isinstance(val, str):
        val = val.strip()
        if val.startswith('[') and val.endswith(']'):
            try:
                return ast.literal_eval(val)
            except:
                return [val]  
        else:
            return [val]
    
    return [val]
```

```python
upn_drop_duplicate['item_code_normalized'] = upn_drop_duplicate['item_code_normalized'].apply(str_to_list)
upn_drop_duplicate['topics'] = upn_drop_duplicate['topics'].apply(str_to_list)
```

```python
def drop_book_duplicates_combined(df):
    df = df.copy()
    keys = ['title', 'authors', 'isbn_issn', 'first_author_last_name']
    
    for col in keys:
        if col in df.columns:
            df[f'{col}_tmp'] = df[col].astype(str).str.lower().str.strip()
            df[f'{col}_tmp'] = df[f'{col}_tmp'].replace(['nan', 'none', '', '__unknown_author__'], 'unknown')

    # menghitung jumlah kolom yang tidak kosong
    df['info_count'] = df.notna().sum(axis=1)

    mask_unknown = df['authors'].isna()
    # Agar data yang tidak ada nama penulis berada di bawah
    df.loc[mask_unknown, 'info_count'] -= 10

    if 'authors' in df.columns:
        mask_unknown = (df['authors'].isna()) | \
            (df['authors'].astype(str).str.lower().str.contains('unknown', na=False))
        df.loc[mask_unknown, 'info_count'] -= 10

    tmp_keys = [f'{c}_tmp' for c in keys if f'{c}_tmp' in df.columns]
    # Pengurutan data menggunakan skor data yang tidak kosong
    df = df.sort_values(by=tmp_keys + ['info_count'], ascending=[True] * len(tmp_keys) + [False])

    # Fungsi untuk menggabungkan data list pada kolom kode_buku dan topik
    def combine_lists(series):
        combined = []
        for val in series:
            if isinstance(val, list):
                combined.extend(val)
            elif pd.notna(val):
                combined.append(val)
        seen = set()
        result = []
        for x in combined:
            if x not in seen:
                seen.add(x)
                result.append(x)
        return result if result else None

    group_cols_list = [
        ['title_tmp', 'isbn_issn_tmp'],
        ['title_tmp', 'first_author_last_name_tmp'],
        ['isbn_issn_tmp', 'authors_tmp'],
        ['isbn_issn_tmp', 'first_author_last_name_tmp']
    ]

    for group_cols in group_cols_list:
        # Penggabungan data kode_buku dan topik
        agg_dict = {
            'item_code_normalized': combine_lists,
            'topics': combine_lists
        }
        # data yang lainnya
        other_cols = [c for c in df.columns if c not in group_cols + ['item_code_normalized', 'topik']]
        for col in other_cols:
            agg_dict[col] = 'first'
        
        # Penggabungan data yang sama untuk menghilangkan duplikasi
        df = df.groupby(group_cols, dropna=False).agg(agg_dict).reset_index()

    # Menghapus kolom sementara
    cols_to_drop = [f'{c}_tmp' for c in keys if f'{c}_tmp' in df.columns] + ['info_count']
    df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
    
    return df
```

```python
upn_drop_duplicate = drop_book_duplicates_combined(upn_drop_duplicate)
show(upn_drop_duplicate[upn_drop_duplicate.duplicated(subset=['title'], keep=False)][['title', 'authors', 'first_author_last_name', 'isbn_issn', 'item_code_normalized']])
```
